```python
#1.Write a Python Program to Find LCM?

def lcm(a, b):
  if a > b:
      greater = a
  else:
      greater = b

  while(True):
      if((greater % a == 0) and (greater % b == 0)):
          out = greater
          break
      greater += 1

  return out

try:
  a,b = input("Enter 2 numbers to find it's L.C.M: a b: ").split()
  print(f"L.C.M of {a}, {b} >> {lcm(int(a),int(b))}")
except Exception as e:
  print("Error: ",str(e))
```

```python
#2. Write a Python Program to Find HCF?
def hcf(a, b):
    if a < b:
        smaller = a
    else:
        smaller = b
    for i in range(1, smaller+1):
        if((a % i == 0) and (b % i == 0)):
            out = i 
    return out

try:
  a,b = input("Enter 2 numbers to find it's H.C.F: a b: ").split()
  print(f"L.C.M of {a}, {b} >> {hcf(int(a),int(b))}")
except Exception as e:
  print("Error: ",str(e))
```

```python
# 3.Write a Python Program to Convert Decimal to Binary, Octal and Hexadecimal?
def dec_to_hexadecimal(n):
  hx = {'10':'A', '11':'B', '12':'C', '13':'D', '14':'E', '15':'F'}
  hx_s = []
  try:
    while n > 0:
      if n%16 == 0:
        s = str(n/16).split('.')
        hx_s.insert(0,'0')
        n = int(s[0])
      if n%16 != 0:
        s = str(n/16).split('.')
        if int(s[1]) > 0:
          hx_s.insert(0,str(int(float("0."+s[1])*16)))
          n = int(s[0])
        else:
          break
    for i in range(len(hx_s)):
      if hx_s[i] in hx.keys():
        hx_s[i] = hx[hx_s[i]]
    output = "".join(hx_s)
    return output
  except Exception as e:
    print(f"Error occured at dec_to_hexa: {str(e)}")

def bin_to_oct(n):
  l = len(n)
  la = 0
  try:
    while l%3 != 0:
      l += 1
      la += 1
    for i in range(la):
      n.insert(0,'0')
    out = []
    for j in range(0,len(n),3):
      out.append(str(int(n[j])*4 + int(n[j+1])*2 + int(n[j+2])*1))
    output = "".join(out)
    return output
  except Exception as e:
    print("Error occured at bin_to_oct: ",str(e))

  
def decm_to_bin(n):
  bin = []
  try:
    while n != 1:
      if n%2 == 0:
        bin.insert(0,'0')
        n = n//2
      else:
        bin.insert(0,'1')
        n = n//2
    bin.insert(0,'1')
    return bin
  except Exception as e:
    print("Error occured at decm_to_bin: ",str(e))
try:
  a = int(input("Enter a Decimal number to find it's Binary, Ocatal, hexadecimal: "))
  print(f"\nEntered decimal number: {a}")
  db = decm_to_bin(a)
  print("Decimal to Binary: ","".join(db))
  bo = bin_to_oct(db)
  print("Decimal to Octal: ",bo)
  dh = dec_to_hexadecimal(a)
  print("Decimal to hexadecimal: ",dh)
except Exception as e:
  print("Error occured: ",e)
```
